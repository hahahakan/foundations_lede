```python
filenames2010 = glob.glob("transcriptstxt/transcripts_conv/transcripts2010/*.txt")
transcripts2010 = [open(filename, encoding="ISO-8859-1").read() for filename in filenames2010]
len(transcripts2010)
```

```python
filedockets = glob.glob("transcriptstxt/transcripts_conv/transcripts2010/*.txt")
case_list2010 = []
for docket in filedockets:
    case = re.findall(r'\d{4}/(\d.*).txt', docket)
    case_list2010.append(case)
```

```python
transcripts2010_cleaned = []
for transcript in transcripts2010:
    beginnings_ends_cleaned = re.sub(r'(Official - Subject to Final Review|Alderson Reporting Company)', '', transcript)
    pagenums_cleaned = re.sub(r'\n\n?\d\d?', '', beginnings_ends_cleaned)
    lines_cleaned = re.sub(r'\n{2,}', ' ', pagenums_cleaned)
    remove_intro = r'(PROCEEDINGS|P R O C E E D I N G S)'
    transcript_1 = re.split(remove_intro, lines_cleaned)
    remove_end = r'(Whereupon)'
    new_transcript = re.split(remove_end, transcript_1[2])
    everything_cleaned = r" ([A-Z .-]{3,}): "
    cleaned = re.split(everything_cleaned, new_transcript[0])
    transcripts2010_cleaned.append(cleaned)
```

```python
len(transcripts2010_cleaned)
```

```python
n = 0 
all2010 = []
for case in case_list2010:
    a2010 = { 'number': case_list2010[n][0], 'text': transcripts2010_cleaned[n]}
    n = n + 1
    all2010.append(a2010)
```

```python
speaker_info = []
for year in all_transcripts_all_years:
    for case in year:
        case = case['text']
        for text in case:
            speaker_info.append(case)
```

```python
num = []
speaker_name = []
speaker_text = []
for year in all_transcripts_all_years:
    for case in year:
        num.append(case['number'])
        for text in case['text'][1::2]:
            speaker_name.append(text)
        for text in case['text'][2::2]:
            speaker_text.append(text)
```

```python
data = []

# from 2010 to 2018
for year in all_transcripts_all_years:

# each case in one year
    for case in year:

        num = []
        speaker_name = []
        speaker_text = []

        speakers = {
            "num": 0,
            "name": 0,
            "text": 0
            }
        speakers["num"] = case['number']
        for text in case['text'][1::2]:
            speaker_name.append(text)
            speakers["name"] = speaker_name
        for text in case['text'][2::2]:
            speaker_text.append(text)
            speakers["text"] = speaker_text
        data.append(speakers)
```

```python
df = pd.DataFrame()
for chief in data:
    chief_df = pd.DataFrame(chief)
    df=pd.concat([df, chief_df])
df.head()
```
